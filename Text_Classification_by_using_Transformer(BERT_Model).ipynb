# Important  libraries

```python
import numpy as np
import pandas as pd
import re
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from datasets import Dataset, DatasetDict
from sklearn.metrics import classification_report
from transformers import EarlyStoppingCallback
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
```

# **Loading dataset**

```python
df = pd.read_csv("/content/Language Detection.csv",encoding='ISO-8859-1')
df.head()
```

# **Data Cleaning**

```python
df_drop = df[['Text','Language']].dropna()
df_drop
```

```python
df_unique = df_drop['Language'].unique()
df_unique
```

```python
valid_languages = ['English','Hindi', 'French', 'Spanish', 'Italian', 'Turkish', 'German']
```

```python
df = df_drop[df_drop['Language'].isin(valid_languages)]
df_unique = df['Language'].unique()
df_unique
```

```python
df
```

```python
def basic_clean_text(text):
   
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    text = text.lower()
    return text

df['Cleaned_Text'] = df['Text'].apply(basic_clean_text)
print(df.head())
```

# **Preprocessing**

```python
train_texts, val_texts, train_labels, val_labels = train_test_split(df['Cleaned_Text'], df['Language'], test_size=0.2, random_state=42)
```

```python
label_encoder = LabelEncoder()
train_labels  = label_encoder.fit_transform(train_labels)
val_labels   = label_encoder.transform(val_labels)

```

```python
train_labels
```

```python
val_labels
```

# **Feature Engineering**

```python
train_dataset = Dataset.from_dict({'text': train_texts.tolist(), 'label': train_labels.tolist()})
val_dataset = Dataset.from_dict({'text': val_texts.tolist(), 'label': val_labels.tolist()})
```

```python
train_dataset
```

```python
val_dataset
```

```python
dataset_dict = DatasetDict({'train': train_dataset,'validation': val_dataset})
dataset_dict
```

 # **Tokenization**

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
def tokenize_function(examples):
    return tokenizer(examples['text'], padding='max_length', truncation=True, max_length=128)

tokenized_datasets = dataset_dict.map(tokenize_function, batched=True)
```

```python
tokenized_datasets
```

# **Using BERT_Model**

```python
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=len(label_encoder.classes_))
```

```python
model
```

# **Model Tuning**

```python

training_args = TrainingArguments(
    output_dir='./results',         
    num_train_epochs=4,            
    per_device_train_batch_size=16,  
    per_device_eval_batch_size=16,   
    warmup_steps=500,               
    weight_decay=0.01,               
    logging_dir='./logs',            
    logging_steps=10,
    evaluation_strategy="epoch",
    learning_rate=2e-5,              
    load_best_model_at_end=True,    
    metric_for_best_model='eval_loss',
    save_strategy="epoch",
    save_total_limit=3,              
    save_steps=100 )
```

```python

trainer = Trainer(
    model=model,                         
    args=training_args,                 
    train_dataset=tokenized_datasets['train'],        
    eval_dataset=tokenized_datasets['validation'],   
    callbacks=[EarlyStoppingCallback(early_stopping_patience=2)])


trainer.train()
```

# **Evaluation**

```python
results = trainer.evaluate()
```

```python
predictions = trainer.predict(tokenized_datasets['validation'])
preds = predictions.predictions.argmax(-1)
```

```python
predicted_labels = label_encoder.inverse_transform(preds)
true_labels = label_encoder.inverse_transform(tokenized_datasets['validation']['label'])
```

```python
print(classification_report(true_labels, predicted_labels))
```
